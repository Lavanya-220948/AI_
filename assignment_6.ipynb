```python
# a recursive function for the Fibonacci sequence
def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)
n=int(input('enter a number :'))
for i in range(n):
  print(fib(i),end='')
```

**Memoization:**

It is technique of storing the results in funtion calling so they don't have to be compute again. They can be used stored results.

**Advantages:**



*   It takes less time for execution by avoiding repeated calculations
*   Improve recusive performance



In the below program we have used memoization so it takes less time than before

```python
# "Memoization" and how it helps recursive functions
memo={}
def fib(n):
  if n in memo:
    return memo[n]
  if n<=1:
    return n
  memo[n]=fib(n-1)+fib(n-2)
  return memo[n]
print(fib(5))
```

```python
#Use reduce() from functools to find the product of a list.
from functools import reduce
nums=[1,2,3,4,6,8]
product=reduce(lambda x,y:x*y,nums)
print(product)
```

```python
#function to check if a number is an "Armstrong Number."
def armstrong(n):
  digits=str(n)
  power=len(digits)
  total=0
  for digit in digits:
    total+=int(digit)**power
  if total==n:
    return "armstrong number"
  else:
    return " not an armstrong number"
n=int(input('enter a number'))
print(armstrong(n))
```

```python
#Create a closure that remembers a base value and adds it to an input.
def outer(base):
  def inner(power):
    return base**power
  return inner
result=outer(2)
print(result(9))
print(result(8))
```

*   G(Global):variables declared outside the function which can be accessible anywhere
*   B(Built-in):python built-in functions

Python always checks in tha order of

```python
# function that converts a decimal number to binary without bin()
def dec_to_bin(num):
  if num==0:
    return 0
  binary=''
  while num:
    binary=str(num%2)+binary
    num=num//2
  return int(binary)
print(dec_to_bin(17))
```

```python
#Create a "Generator" function using yield to produce infinite even numbers.
def even():
  num=0
  while True:
    yield num
    num+=2
gen=even()
for i in range(12):
  print(next(gen))
```

```python
#function to validate a password (min 8 chars, 1 digit, 1 symbol)
import re
pattern=r'^(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$'
password=input('enter a password')
if re.match(pattern,password):
  print("valid password")
else:
  print("invalid password")
```
