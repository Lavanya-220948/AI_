```python
#usage  function that accepts variable-length arguments (*args)
def addition(*args):
  summation=sum(args)
  return summation
addition(24,37,56)
```

```python
#a function that accepts keyword arguments (kwargs) and prints them
def student(**kwargs):
  for i in kwargs:
    print(i,':',kwargs[i])
student(name='LAVANYA',city='vizag')
```

```python
#recursive function to calculate the factorial of a number
def factorial(n):
  if n<=1:
    return 1
  else:
    return n*factorial(n-1)
factorial(7)
```

**Global variable**:



*   A global variable is declared outside all functions
*   It can be accessed from anywhere in the program
**Local Variable:**



*   A local variable is declared inside a function
*   It can only be accessed within that function

```python
#Create a "Filter" function to extract prime numbers from a list.
def prime(n):
  if n<2:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True

nums=[4,5,23,45,67,42,89,93]
print(list(filter(prime,nums)))
```

```python
#use of nonlocal in a nested function.
def airthmetic(a,b):
  sum=a+b
  diff=a-b
  return sum,diff
s,d=airthmetic(5,8)
print('sum :',s)
print('difference: ',d)
```

```python
#increment function
def increment():
  a=10
  def total():
    nonlocal a
    a+=1
    print(a)
  total()
increment()
```

**nonlocal keyword**: This keyword is used inside nested function to modify a variable belongs to outer function
